```python
# Import packages
import plotnine as p9
import numpy as np
import statistics
from scipy import stats
from scipy.stats import chi2_contingency
import missingno as msno
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels as sm
import seaborn as sns
#%matlpotlib inline

#import the file
dataf = pd.read_csv('churn_clean207.csv',dtype={'locationid':np.int64}, index_col=0)

#create scatterplot of Bandwidth GB Year vs. Monthly Charge 
plt.scatter(dataf.Bandwidth_GB_Year[0:50], dataf.MonthlyCharge[0:50])
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

red_patch = mpatches.Patch(color='red', label='The red data')
blue_patch = mpatches.Patch(color='blue', label='The blue data')

plt.legend(handles=[red_patch, blue_patch])
plt.title('Bandwidth GB Per Year vs. Monthly Charge')
plt.xlabel('Bandwidth GB Year')
plt.ylabel('Monthly Charge')
```

```python
#import mpatches for legend
import matplotlib.patches as mpatches

#assign both to a variable
bivar = dataf[['StreamingMovies','StreamingTV']]

#plot histogram 
plt.hist(bivar, bins = 2)

#name labels and title
plt.xlabel('StreamingMovies          StreamingTV')
plt.ylabel('# of Subscriptions')
plt.title('Streaming TV vs Streaming Movies ')

#label legend
red_patch = mpatches.Patch(color='blue', label='Yes')
blue_patch = mpatches.Patch(color='orange', label='No')
plt.legend(handles=[red_patch, blue_patch])
plt.tight_layout()
```
